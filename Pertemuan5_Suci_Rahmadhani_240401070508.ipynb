Visualisasi data merupakan salah satu tahap penting dalam Data Science karena membantu mengubah data menjadi informasi yang mudah dipahami. Berbagai jenis grafik seperti bar chart, histogram, boxplot, dan scatter plot memiliki fungsi yang berbeda sesuai dengan karakteristik data yang dianalisis.

Pada praktikum ini digunakan dataset Heart Disease UCI yang berisi informasi mengenai kondisi pasien, seperti usia, jenis kelamin, tekanan darah, kadar kolesterol, serta status penyakit jantung. Hasil visualisasi kemudian disusun menjadi sebuah dashboard statis untuk mempermudah proses analisis.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv("heart_disease_uci.csv")

print("Shape Dataset :", df.shape)

print("\nTipe Data")
display(df.dtypes)

print("\nLima Data Pertama")
display(df.head())

print("\nStatistik Deskriptif")
display(df.describe().round(2))
```

```python
print("Missing Values")

display(df.isnull().sum())
```

### Interpretasi

Dataset Heart Disease UCI terdiri dari 920 baris dan 16 kolom yang berisi informasi karakteristik pasien serta hasil pemeriksaan medis. Variabel yang tersedia meliputi usia, jenis kelamin, tekanan darah, kadar kolesterol, denyut jantung maksimum, dan status penyakit jantung. Eksplorasi awal dilakukan untuk memahami struktur data sebelum membuat visualisasi pada tahap berikutnya. Selain itu, hasil pemeriksaan menunjukkan bahwa dataset masih memiliki beberapa missing values pada beberapa kolom, sehingga hal tersebut perlu diperhatikan apabila dataset akan digunakan untuk analisis statistik lanjutan maupun machine learning. Namun pada praktikum ini fokus utama adalah membuat visualisasi data sehingga data tetap digunakan sesuai kondisi aslinya.

# Visualisasi 1: Distribusi Status Penyakit Jantung

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="num",
    palette="Set2"
)

plt.title("Jumlah Pasien Berdasarkan Status Penyakit Jantung")
plt.xlabel("Status Penyakit Jantung")
plt.ylabel("Jumlah Pasien")

plt.show()
```

What?

Grafik menunjukkan bahwa kategori status penyakit jantung **0** memiliki jumlah pasien paling banyak, sedangkan kategori **4** merupakan yang paling sedikit.

So What?

Distribusi ini menunjukkan bahwa dataset tidak memiliki jumlah data yang seimbang pada setiap kategori sehingga perlu diperhatikan apabila digunakan untuk pemodelan machine learning.

Now What?

Analisis berikutnya dapat membandingkan status penyakit jantung dengan usia, tekanan darah, dan kadar kolesterol untuk mengetahui faktor-faktor yang paling berpengaruh.

# Visualisasi 2: Distribusi Usia Pasien

```python
plt.figure(figsize=(7,4))

sns.histplot(
    data=df,
    x="age",
    bins=20,
    kde=True,
    color="steelblue"
)

plt.axvline(
    df["age"].mean(),
    color="red",
    linestyle="--",
    label=f"Mean = {df['age'].mean():.1f}"
)

plt.axvline(
    df["age"].median(),
    color="orange",
    linestyle="-.",
    label=f"Median = {df['age'].median():.1f}"
)

plt.title("Distribusi Usia Pasien")

plt.xlabel("Usia")

plt.ylabel("Frekuensi")

plt.legend()

plt.show()
```

### What?

Histogram menunjukkan bahwa usia pasien pada dataset memiliki rentang yang cukup luas, yaitu mulai dari **nilai minimum hingga maksimum** tahun. Rata-rata usia pasien adalah sekitar **53.51 tahun**, sedangkan nilai median berada di sekitar **54.00 tahun**. Posisi mean dan median yang relatif berdekatan menunjukkan bahwa distribusi usia tidak terlalu menceng ke salah satu sisi.

### So What?

Distribusi ini menunjukkan bahwa sebagian besar data berasal dari kelompok usia dewasa hingga lanjut usia. Informasi tersebut penting karena usia merupakan salah satu faktor risiko yang sering dikaitkan dengan penyakit jantung dan dapat memengaruhi hasil analisis berikutnya.

### Now What?

Tahap selanjutnya dapat dilakukan dengan membandingkan distribusi usia berdasarkan status penyakit jantung (`num`) untuk mengetahui apakah kelompok usia tertentu memiliki kecenderungan risiko penyakit jantung yang lebih tinggi.

# Visualisasi 3: Kolesterol Berdasarkan Status Penyakit Jantung

```python
plt.figure(figsize=(7,5))

sns.boxplot(
    data=df,
    x="num",
    y="chol",
    palette="Set2"
)

plt.title("Distribusi Kolesterol Berdasarkan Status Penyakit Jantung")

plt.xlabel("Status Penyakit Jantung")

plt.ylabel("Kolesterol")

plt.show()
```

### What?

Boxplot memperlihatkan distribusi kadar kolesterol pada setiap kategori status penyakit jantung. Terlihat adanya variasi nilai dan beberapa outlier pada masing-masing kelompok.

### So What?

Visualisasi ini membantu membandingkan apakah pasien dengan tingkat penyakit jantung tertentu memiliki kecenderungan kadar kolesterol yang lebih tinggi dibandingkan kelompok lainnya.

### Now What?

Analisis statistik lanjutan dapat dilakukan untuk mengetahui apakah perbedaan kadar kolesterol antar kelompok signifikan secara statistik.

# Visualisasi 4: Hubungan Usia dan Denyut Jantung Maksimum

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    data=df,
    x="age",
    y="thalch",
    hue="sex",
    palette="Set2"
)

sns.regplot(
    data=df,
    x="age",
    y="thalch",
    scatter=False,
    color="black"
)

plt.title("Hubungan Usia dan Denyut Jantung Maksimum")

plt.xlabel("Usia")

plt.ylabel("Maximum Heart Rate")

plt.show()
```

What?

Scatter plot menunjukkan kecenderungan hubungan negatif antara usia dan denyut jantung maksimum.

So What?

Semakin bertambah usia pasien, denyut jantung maksimum cenderung menurun.

Now What?

Hubungan ini dapat dianalisis lebih lanjut menggunakan korelasi Pearson maupun regresi linear.

# Dashboard Visualisasi Statis

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# ===================================================
# Grafik 1 - Bar Chart
# ===================================================

sns.countplot(
    data=df,
    x="num",
    palette="Set2",
    ax=axes[0,0]
)

axes[0,0].set_title("Status Penyakit Jantung")
axes[0,0].set_xlabel("Status")
axes[0,0].set_ylabel("Jumlah")

# ===================================================
# Grafik 2 - Histogram
# ===================================================

sns.histplot(
    data=df,
    x="age",
    bins=20,
    kde=True,
    color="steelblue",
    ax=axes[0,1]
)

axes[0,1].set_title("Distribusi Usia")
axes[0,1].set_xlabel("Usia")

# ===================================================
# Grafik 3 - Boxplot
# ===================================================

sns.boxplot(
    data=df,
    x="num",
    y="chol",
    palette="Set2",
    ax=axes[1,0]
)

axes[1,0].set_title("Kolesterol Berdasarkan Status Penyakit")

# ===================================================
# Grafik 4 - Scatter Plot
# ===================================================

sns.scatterplot(
    data=df,
    x="age",
    y="thalch",
    hue="sex",
    palette="Set2",
    ax=axes[1,1]
)

axes[1,1].set_title("Usia vs Maximum Heart Rate")

plt.suptitle(
    "Dashboard Visualisasi Heart Disease UCI Dataset",
    fontsize=16,
    fontweight="bold"
)

plt.tight_layout()

plt.show()
```

### Interpretasi Dashboard

Dashboard ini menggabungkan empat jenis visualisasi yang saling melengkapi sehingga karakteristik dataset dapat dipahami dengan lebih mudah. Bar chart menunjukkan distribusi status penyakit jantung, histogram menggambarkan persebaran usia pasien, boxplot membandingkan kadar kolesterol pada setiap kategori penyakit, sedangkan scatter plot memperlihatkan hubungan antara usia dan denyut jantung maksimum.

Dengan menampilkan seluruh grafik dalam satu dashboard, proses eksplorasi data menjadi lebih efisien karena informasi penting dapat diamati secara bersamaan. Dashboard seperti ini sering digunakan sebagai tahap awal sebelum melakukan analisis statistik lanjutan maupun membangun model machine learning.

```python
plt.tight_layout(rect=[0,0,1,0.96])
fig.savefig(
    "dashboard_heart_disease.png",
    dpi=300,
    bbox_inches="tight"
)

print("Dashboard berhasil disimpan sebagai dashboard_heart_disease.png")
```
